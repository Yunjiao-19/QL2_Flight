## Narrowing down Time Frame of Top5 Short Dataset

```python
#Splitting the departure_date into new columns of year/month
    
try:
    ctx= snowflake.connector.connect(user='CMU',
                                     password="ql2_analytics",
                                     account="zk56677",
                                     database='QL2_PROD',
                                     schema='public',
                                     warehouse='CMU_WAREHOUSE',
                                     role='SYSADMIN' )
    cursor = ctx.cursor()
    cur = cursor.execute(""" 
    WITH df1 as(
     SELECT 
         MARKET, SITE, CXR, QTS, QL2_QTS, DEPARTURE_DATE, QL2_DDATE, DEPARTURE_STOPS,FARE_TYPE, FARE, 
         FLIGHT_NUMBER, DEPARTURE_TIME, ROUTE, YEAR_MONTH, DURATION_FLIGHT,
         LEFT(DURATION_FLIGHT,CHARINDEX(':',DURATION_FLIGHT)-1) AS DURATION_HRS, 
         CAST(DURATION_HRS AS FLOAT)DURATION_HRSINT
     FROM CAPSTONE_RAW_AIRFARE_V 
     )
     SELECT *, EXTRACT(MONTH FROM QTS) AS MONTH, EXTRACT(YEAR FROM QTS) AS YEAR 
             FROM df1 WHERE (
                 DURATION_HRSINT <=3 
                 AND DEPARTURE_STOPS = 1 
                 AND (FARE_TYPE = 'ECONOMY' 
                 OR FARE_TYPE = 'Economy') 
          ) ;
 
    """)

    df_top5short = cur.fetch_pandas_all()
    
except Exception as e:
    print("failed connecting to QL2_PROD",e)
```

```python
#CHECKING DISTRIBUTION OF DATA BY MONTH/YEAR
    
try:
    ctx= snowflake.connector.connect(user='CMU',
                                     password="ql2_analytics",
                                     account="zk56677",
                                     database='QL2_PROD',
                                     schema='public',
                                     warehouse='CMU_WAREHOUSE',
                                     role='SYSADMIN' )
    cursor = ctx.cursor()
    cur = cursor.execute(""" 
    WITH df1 as(
     SELECT 
         MARKET, SITE, CXR, QTS, QL2_QTS, DEPARTURE_DATE, QL2_DDATE, DEPARTURE_STOPS,FARE_TYPE, FARE, 
         FLIGHT_NUMBER, DEPARTURE_TIME, ROUTE, YEAR_MONTH, DURATION_FLIGHT,
         LEFT(DURATION_FLIGHT,CHARINDEX(':',DURATION_FLIGHT)-1) AS DURATION_HRS, 
         CAST(DURATION_HRS AS FLOAT)DURATION_HRSINT
     FROM CAPSTONE_RAW_AIRFARE_V 
     ), 
     df2 as(
     SELECT *, EXTRACT(MONTH FROM QTS) AS MONTH, EXTRACT(YEAR FROM QTS) AS YEAR 
             FROM df1 WHERE (
                 DURATION_HRSINT <=3 
                 AND DEPARTURE_STOPS = 1 
                 AND (FARE_TYPE = 'ECONOMY' 
                 OR FARE_TYPE = 'Economy') 
          )
    )
    SELECT MONTH,YEAR, COUNT(*) 
    FROM df2 
    GROUP BY MONTH,YEAR
    ORDER BY max(QTS) ASC
    
    ;
 
    """)

    df_top5short = cur.fetch_pandas_all()
    
except Exception as e:
    print("failed connecting to QL2_PROD",e)
```

# Clean data

```python
a = df_top5short_final
```

```python
# check uniqueness of each column
#a['CURRENCY'].unique()
#a['RDATE'].unique()
```

```python
import time
import datetime as dt

# transforming type of DATE and QTS
a['tDATE'] = pd.to_datetime(a['DEPARTURE_DATE'])
a['tqts'] = pd.to_datetime(a['QTS'])
# calculating time till Departure
a['re'] = a['tDATE'] - a['tqts']
a
```

```python
# calculate days till departure
a['DaysTillDeparture'] = a['re'].dt.components.days
a
```

```python
# clean the data/type of DURATION

# first convert to string
def convert_to_string (x):
    return str(x).split('\n')[0].split(' ')[-1]+':00'
a['DURATION_FLIGHT'] = a['DURATION_FLIGHT'].apply(convert_to_string)
a['DURATION_FLIGHT']
```

```python
# convert DURATION from string to timedelta object
a['DURATION_FLIGHT']= pd.to_timedelta(a['DURATION_FLIGHT'])
a['DURATION_FLIGHT']
```

```python
# convert DURATION from timedelta object to total seconds in float type
a['DURATION_FLIGHT'] = a['DURATION_FLIGHT'].apply(lambda x: x.total_seconds())
a['DURATION_FLIGHT']
```

```python
a
```

## Adding Features: Holiday, Seasonal, Weekend/Weekday, Pandemic

```python
#pip install holidays
```

```python
import holidays

# holidays of US
us_holidays = holidays.UnitedStates()
```

```python
b = a

# adding pandemic features
b['Pandemic'] = (a['DEPARTURE_DATE'] > dt.date(2020,3,20)).apply(lambda x: int(x))
```

```python
us_holidays.get(dt.date(2019,11,28).strftime("%m-%d-%Y"))
```

```python
# adding holiday features

## check if a date is a US holiday
def checkHoliday(date):
    holiday_name = us_holidays.get(date.strftime("%m-%d-%Y"))
    if holiday_name != None:
        return holiday_name
    else:
        return "None"

b['Holidays'] = b['DEPARTURE_DATE'].apply(lambda x: checkHoliday(x))
```

# Implementing Linear Regression

```python
from sklearn.linear_model import LinearRegression
# drop FARE because it's response variable
# drop SITE, CURRENCY because they only have 1 single value throughout df, so they are not helpful
# drop QTS because we have tqts
X = b[['DaysTillDeparture','FARE_TYPE', 'MARKET','CXR', 'DEPARTURE_STOPS', 'DURATION_FLIGHT', 'Pandemic', 'Holidays']]
X
```

```python
# response variable
y = b['FARE']
y
```

```python
# convert categorical variables to dummy variables
X = pd.get_dummies(X, columns=['FARE_TYPE', 'MARKET','CXR', 'Pandemic', 'Holidays'])
X
```

###  statistical analysis of feeding all data for linear regression

```python
# fit linear regression
reg = LinearRegression().fit(X, y)
```

```python
# linear regression summary
import statsmodels.api as sm
mod = sm.OLS(y, X)
res = mod.fit()
print(res.summary())
```

### Based on the p-values given above, DaysTillDeparture, DURATION_FLIGHT, MARKET, CXR, Holidays_Christmas Day, Holidays_Independence Day , Holidays_Juneteenth National Independence Day , Holidays_Martin Luther King Jr. Day, Holidays_Veterans Day , Holidays_None, Holidays_Washington's Birthday are important factors as they have p-values lower than 0.05, indicating strong evidence of relationship with response variable.
